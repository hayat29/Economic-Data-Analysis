```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('Cost_of_Living_Index_2022.csv')

# Print available columns for debugging
print("Available columns:", df.columns)

# Data Cleaning Function
def clean_data(df):
    """Cleans dataset and ensures necessary features are present."""
    
    # Convert numerical columns safely
    numeric_cols = ['Cost of Living Index', 'Rent Index', 'Cost of Living Plus Rent Index',
                    'Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index']
    
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    return df

df = clean_data(df)

# Function to plot cost of living across countries
def plot_cost_of_living(df):
    """Plots the top 15 countries with the highest cost of living."""
    if 'Country' not in df.columns or 'Cost of Living Index' not in df.columns:
        print("Skipping cost of living analysis due to missing data.")
        return
    
    top_countries = df.nlargest(15, 'Cost of Living Index')

    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_countries['Cost of Living Index'], y=top_countries['Country'], palette='coolwarm')
    plt.title('Top 15 Countries with Highest Cost of Living Index')
    plt.xlabel('Cost of Living Index')
    plt.ylabel('Country')
    plt.tight_layout()
    plt.show()

plot_cost_of_living(df)

# Function to visualize index comparisons across regions
def plot_index_comparison(df):
    """Plots various indices (Rent, Groceries, Restaurant, Purchasing Power) across top cost-of-living countries."""
    index_cols = ['Rent Index', 'Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index']
    
    if not {'Country'}.issubset(df.columns) or not all(col in df.columns for col in index_cols):
        print("Skipping index comparison due to missing data.")
        return
    
    top_countries = df.nlargest(15, 'Cost of Living Index')
    top_countries = top_countries.set_index('Country')

    plt.figure(figsize=(15, 6))
    top_countries[index_cols].plot(kind='bar', figsize=(15, 6), colormap='viridis')
    plt.title('Comparison of Living Indices Across Countries')
    plt.xlabel('Country')
    plt.ylabel('Index Value')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_index_comparison(df)

# Function to analyze correlation between different cost indices
def plot_correlation_matrix(df):
    """Plots correlation heatmap for different cost indices."""
    index_cols = ['Cost of Living Index', 'Rent Index', 'Cost of Living Plus Rent Index',
                  'Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index']

    if not all(col in df.columns for col in index_cols):
        print("Skipping correlation analysis due to missing data.")
        return

    corr_matrix = df[index_cols].corr()

    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Matrix of Cost Indices')
    plt.tight_layout()
    plt.show()

plot_correlation_matrix(df)

# Summary Report
def generate_summary(df):
    """Generates key insights summary based on cost of living data."""
    if 'Country' not in df.columns or 'Cost of Living Index' not in df.columns:
        print("Skipping summary due to missing data.")
        return

    highest_cost_country = df.nlargest(1, 'Cost of Living Index')['Country'].values[0]
    lowest_cost_country = df.nsmallest(1, 'Cost of Living Index')['Country'].values[0]

    avg_rent_index = df['Rent Index'].mean()
    avg_grocery_index = df['Groceries Index'].mean()
    avg_purchasing_power = df['Local Purchasing Power Index'].mean()

    summary = f"""
    🏆 **Key Insights**
    - Country with the highest cost of living: **{highest_cost_country}**
    - Country with the lowest cost of living: **{lowest_cost_country}**
    - Average Rent Index: **{avg_rent_index:.2f}**
    - Average Groceries Index: **{avg_grocery_index:.2f}**
    - Average Local Purchasing Power Index: **{avg_purchasing_power:.2f}**

    📊 **Observations**
    - Countries with high rent index often have high cost of living.
    - Purchasing power tends to be **inversely** correlated with cost of living.
    - Some regions experience lower cost of living but significantly higher restaurant prices.
    """
    print(summary)

generate_summary(df)
```
